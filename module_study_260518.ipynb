### 학습 되돌아 보기(2)

```python
import math
import random

def getarea(r):
    return math.pi * (r ** 2)

radius = round(random.random() * 10, 2)
area = getarea(radius)

#output
print(f'원의 반지름: {radius}')
print(f'원주율 pi: {math.pi:.4f}')
print(f'반지름 {radius}인 원의 면적은 {area:.2f}')
```

### 학습 되돌아 보기(3)

```python
import math
import statistics

def run_statistics():
    print(f"1! = {math.factorial(1)}")
    print(f"6! = {math.factorial(6)}")
    print(f"11! = {math.factorial(11)}")
    print(f"16! = {math.factorial(16)}")
    print()

    # Data
    data = [80, 99, 77, 65, 92, 74, 82]
    print(data)

    # Calculate
    median_val = statistics.median(data)
    mean_val = statistics.mean(data)
    variance_val = statistics.variance(data)
    stdev_val = statistics.stdev(data)

    # Output 
    print(f"중앙 값: {median_val:.2f}")
    print(f"평균: {mean_val:.2f}")
    print(f"분산: {variance_val:.2f}")
    print(f"표준편차: {stdev_val:.2f}")

if __name__ == '__main__':
    run_statistics()
```

### 이미지 파일 복사하기

```python
import os

def copy_image_file(image_file):
    # 파일명과 확장자 분리 후 '_copy' 결합
    name, ext = os.path.splitext(image_file)
    copy_file = f"{name}_copy{ext}"
    
    # 원본 읽어서 복사본 쓰기
    with open(image_file, 'rb') as f:
        content = f.read()
        
    with open(copy_file, 'wb') as F:
        F.write(content)
        
    print(copy_file)
```
