```python
# while문 연습 : 나무 찍기

tree_hit = 1
while tree_hit < 10:
    print(f'나무를 {tree_hit}번 찍었습니다.')
    tree_hit += 1
print("나무가 넘어갑니다.")
```

```python
# while문 연습 : 입력 값에 따라 반복문 종료하기
# 사용자로부터 숫자를 하나씩 입력 받아서 총합과 평균을 출력
# 사용자가 0을 입력하면 숫자 입력을 멈춤


sum = count = 0
num = int(input("숫자 입력 (0 : 끝) :"))
while num != 0:
    sum += num
    count += 1
    num = int(input("숫자 입력 (0 : 끝) :"))

# 0 으로 나누기를 하면 ZeroDivisionError 발생 -> 예외처리
if count != 0:
    average = sum / count
else: 
    average = 0
    
print(f">>총합 : {sum}, 평균 : {average:.2f}")
```

```python
# 무한루프와 분기문 break 활용
sum = count = 0
while True:
    num = int(input("숫자 입력 (0 : 끝) :"))
    if num ==0:
        break
    sum += num
    count +=1
    

# 0 으로 나누기를 하면 ZeroDivisionError 발생 -> 예외처리
if count != 0:
    average = sum / count
else: 
    average = 0
    
print(f">>총합 : {sum}, 평균 : {average:.2f}")
```

```python
# for 문 연습
sum = 0


for i in 1.1, 2.5, 3.6, 4.2, 5.4:
    sum += i
    print(i, sum)

print(f'합 : {sum}, 평균 : {sum / 5:.2f}')
```

```python
# 과제 5개의 숫자 입력하여 더하기
print('5개의 숫자를 입력하세요.')

sum = 0

for i in range(5):
    num = int(input())
    sum += num

print(f'5개의 숫자의 합은 {sum}입니다.')
```

```python
# 1부터 N까지의 합 구하기 

number = int(input("숫자를 입력하세요 :"))

sum = 0
for num in range(1, number + 1): 
    sum += num
    
print(f'1부터 {number}까지의 합은 {sum}입니다.')
```

```python
# 숫자를 입력받고 그 숫자에 해당하는 구구단 출력

num = int(input('몇 단을 출력할까? (1~9까지 숫자 입력)'))

if 1 <= num <= 9 :
    for i in range(1,10):
        print(f'{num} x {i} = {num * i}')

else : 
    print('구구단을 출력할 수 없어!')
```

```python
#숫자를 입력받고 그 수의 팩토리얼을 출력

fac = 1
num = int(input('몇 팩토리얼을 출력할까? :'))

for i in range(2, num + 1) :
    fac *= i

print(f'{num}! = {fac}')
```
