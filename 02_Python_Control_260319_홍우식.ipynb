```python
# 3개의 숫자 중 최댓값 구하기

# a, b, c = input("세 개의 숫자를 입력하세요"). split()
# a = int(a)
# b = int(b)
# c = int(c)
# print(type(a))

a, b, c = map(int, input("세 개의 숫자를 입력하세요"). split())

max = a 
if max < b: max = b
if max < c: max = c
print(f'가장 큰 수는 {max}입니다.')
```

```python
age1 = int(input('첫번째 사람의 나이:'))
age2 = int(input('두번째 사람의 나이:'))

if age1 > age2:
    diff = age1 - age2
else:
    diff = age2 - age1

print('>> 나이차이 :', diff)
```

```python
# 입력된 숫자가 음성, 0, 양수인지 출력
num = int(input('숫자를 입력하세요 :'))
if num > 0:
    print('>> 양수입니다')
elif num == 0:
    print('>> 0입니다')
else:
    print('>> 음수입니다')
```

```python
# 입력된 점수에 따라 등급을 출력
score = int(input('점수를 입력하세요 :'))

if score >= 90: grade = 'A'

elif score >= 80: grade = 'B'

else: grade = 'F'


print(f'>> 등급은 {grade}입니다.')
```

```python
# 숫자를 입력 받아서 양수인 경우에만 짝수인지 홀수인지 출력한다. 양수가 아니면, 양수가 아니라고 출력한다.

num = int(input('숫자를 입력하세요 :'))

if num > 0:
    if num % 2 == 0:
        print('짝수입니다.')
    else:
        print('홀수입니다.')
else:
    print('양수가 아닙니다.')
```
