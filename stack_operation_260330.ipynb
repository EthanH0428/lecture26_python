```python
stack = [] 
capacity = 5

def isFull():
    if len(stack) == capacity:
        return True
    else:
        return False

def isEmpty():
    if len(stack) == 0:
        return True
    else:
        return False

def push(data):
    if isFull():
        print('Stack이 차 있어서 더 이상 추가할 수 없다.')
    else:
        stack.append(data)

def pop():
    if isEmpty():
        print()
        return -1
    else:
        return stack.pop()

def peek():
    if isEmpty(): 
        print('Stack이 비어있어서 더 이상 꺼내볼 수 없다.')
        return -1
    else:
        return stack[-1]
        
        
    

#=======================================================
print(f'[[ 정수형 스택 연산 실습 (용량 : {capacity})]]')


while True:
    menu = int(input('> 메뉴 선택: '))
    if menu == 0: 
        break
    elif menu == 1: # Push
        data = int(input('데이터 입력: '))
        push(data)
    elif menu == 2: # pop
        data = pop()
        print('[Pop] 가져온 데이터 :', data)
    elif menu == 3: # peek
        data = peek()
        print('top 데이터 확인 :', data)
        continue
        
    print('> 현재 스택 상태', stack)
print('[[ 정수형 스택 연산 실습 종료]]')
```
