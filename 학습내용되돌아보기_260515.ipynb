### 학습 내용 되돌아 보기(2)

```python
try:
    m = int(input('월(month) 입력 > '))
    if not (1 <= m <= 12):
        # 빈칸: 예외 강제 발생
        raise Exception('1~12 사이의 정수를 입력하세요.')
except Exception as e:
    print(f'예외 발생: {e}')
else:
    print(f'{m}월: 잘 입력했습니다.')
```

```python
def divide(x, y):
    try:
        result = x / y
    except ZeroDivisionError:
        print("0으로는 나눌 수 없습니다.")
    else:
        print(f"결과: {result}")

# 함수 호출
divide(3.2, 2)
divide(5.4, 0)
```
