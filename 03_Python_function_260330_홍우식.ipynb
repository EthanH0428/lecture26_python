```python
# num에 해당하는 구구단 출력 

def gugudan(num):
    print(f'[구구단 {num}단 ]')
    for i in range(1,10):
        print(f'{num} x {i} = {num*i}')
gugudan(5)
```

```python
number = int(input('단 입력'))
gugudan(number)
```

```python
# 2단부터 9단까지 출력하기
for i in range(2, 10):
    gugudan(i)
    print()
```

```python
# return 값 확인
def add_and_mul(a, b):
    return a + b, a * b

result = add_and_mul(3, 4)
print(result)
result1, result2 = add_and_mul(3, 4)
print(result1, result2)
```

```python
# 지역변수, 전역변수
sum = 10
def test(sum, num):
    sum += num
    return sum
    
sum = test(sum, 5)
print(sum)
```
