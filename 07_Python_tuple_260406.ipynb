### 튜플 만들기

```python
t1 = ()
t2 = (1,)
t3 = 1, 2, 3
t4 = 1, 

print(t3)
x, y, z = t3
print(x, y, z)
```

### 튜플의 활용

```python
tuple = (10, 20, 30)
print("현재 튜플:", tuple)
```

```python
x, y = 10, 20
x, y = y, x
print(x, y)
```

```python
def add_and_mul(a, b):
    return a+b, a*b

result = add_and_mul(3, 5)
print(result)
```

```python
fruits = ["apple", "banana", "strawberry"]

print("blueberry" in fruits)
print("apple" in fruits)
```

```python
snack = ("쿠키", "초코")
print("삭제 전:", snack)

del snack

print("상자 자체를 del로 삭제하는 건 성공!")
```

```python
scores = (85, 92, 78, 92, 88)

print(f"전체 과목 수: {len(scores)}")
print(f"최고 점수: {max(scores)}")
print(f"최저 점수: {min(scores)}")

count_92 = scores.count(92)
print(f"92점의 개수: {count_92}")

index_78 = scores.index(78)
print(f"78점의 위치: {index_78}번 인덱스")
```
