```python
#구구단

num = int(input('출력할 단 :'))

print(f'[구구단 {num}단 ]')
for i in range(1,10):
    print(f'{num} x {i} = {num*i}')
```

```python
for num in range(2,10):
    # num에 해당하는 구구단 출력
    print(f'[구구단 {num}단 ]')
    for i in range(1,10):
        print(f'{num} x {i} = {num*i}')
```
